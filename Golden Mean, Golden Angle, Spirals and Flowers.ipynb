## Spirals, Radial Plots, Flowers and Mounds created with Python
Mathematics and Jupyter notebook can be used together to create interesting shapes, spirals and flowers. We are using the concepts of Parametric equations and Golden Ratio to create these interesting flowers and spirals.
You can see the spirals showing the golden ratio.
This notebook is a tribute to the beautiful shapes and flowers 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
r = np.linspace(0, 10, num=100)
df = pd.DataFrame({'r': r, 'slow': r, 'medium': 2 * r, 'fast': 4 * r})


df = pd.melt(df, id_vars=['r'], var_name='speed', value_name='theta')


g = sns.FacetGrid(df, col="speed", hue="speed",
                  subplot_kws=dict(projection='polar'), height=4.5,
                  sharex=False, sharey=False, despine=False)
g.map(sns.scatterplot, "theta", "r");
plt.savefig('radial.png')
```

```python
r = np.linspace(0, 50, num=500)
df = pd.DataFrame({'r': r, 'slow': r, 'medium': 2 * r, 'fast': 4 * r})
df = pd.melt(df, id_vars=['r'], var_name='speed', value_name='theta')
g = sns.FacetGrid(df, col="speed", hue="speed",
                  subplot_kws=dict(projection='polar'), height=4.5,
                  sharex=False, sharey=False, despine=False)
g.map(sns.scatterplot, "theta", "r");
```

# Using Parametric Equations
**We are familiar with equations like y=mx+b where y is expressed in terms of x. To create radial diagrams we have to express x and y in terms of a new variable. That's why we will use a parametric equation where we can have two y's from a single value of x, which is not allowed in an equation of a straight line, e.g. a circle. The equations x=np.sin(t) and y=np.cos(t) still define a function as we are trying to create a spiral.**

```python
points=500
angle=np.pi*(3-np.sqrt(5))
t=np.arange(1,501)*angle
x=np.sin(t)
y=np.cos(t)
df1=pd.DataFrame({'t':t,'x':x,'y':y})
df1.head()
fig=plt.figure(figsize=(15,15))
fig1=plt.scatter(df1['x']*df1['t'],df1['y']*df1['t'],alpha=0.5,c='red',marker='o')
fig2=plt.axis('off')
```

# Lets use the Golden Ratio as our angle

# GOLDEN RATIO
Throughout history, the ratio for length to width of rectangles of 1.61803 39887 49894 84820 has been considered the most pleasing to the eye. This ratio was named the golden ratio by the Greeks. In the world of mathematics, the numeric value is called "phi", named for the Greek sculptor Phidias.
- You can find the Golden Ratio when you divide a line into two parts and the longer part (a) divided by the smaller part (b) is equal to the sum of (a) + (b) divided by (a), which both equal 1.618. This formula can help you when creating shapes, logos, layouts, and more.
![phi](https://kids.kiddle.co/images/math/7/1/f/71f71504d6761bb495a15c9b410b4093.png)

![Seashell](https://p0.pxfuel.com/preview/549/362/542/fibonacci-spiral-seashell-spiritual.jpg)

```python
plt.style.use('dark_background')
points=1000
angle=(1+np.sqrt(5))/2
t=np.arange(1,1001)*angle
x=np.sin(t/99)
y=np.cos(t/99)
df1=pd.DataFrame({'t':t,'x':x,'y':y})
df1.head()
fig=plt.figure(figsize=(15,15))
fig1=plt.scatter(df1['x']*df1['t'],df1['y']*df1['t'],alpha=0.5,c='cyan',marker='o')
fig2=plt.axis('off')
plt.savefig('spiral1.png')
```
